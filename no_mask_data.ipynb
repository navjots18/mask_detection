```python
import uuid
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
import cv2
import requests
from io import BytesIO
```

```python
url='https://media.istockphoto.com/photos/portraits-of-faces-picture-id473223214'
rel = requests.get(url)
rel.status_code
```

```python
f = open("image.jpg","wb")
f.write(rel.content)
f.close()
```

```python
frame = cv2.imread("image.jpg")
```

```python
plt.imshow(frame)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
faces= classifier.detectMultiScale(frame)
```

```python
pil = Image.open("image.jpg") 
arr= np.array(pil)
```

```python

for face in faces:
    x,y,h,w = face
    img = Image.fromarray(arr[y:y+h,x:x+w,], 'RGB')
    img = img.resize((40,40))
    plt.figure()
    plt.imshow(arr[y:y+h,x:x+w,])
   
```

```python
i=1
for face in faces:
    x,y,h,w = face
    img = Image.fromarray(arr[y:y+h,x:x+w,], 'RGB')
    a=str(uuid.uuid4())
    img = img.resize((40,40))
    c=input("save "+str(i)+"?(y/n)")
    i+=1
    if(c=='y'):
        img = img.save("no_mask/"+a +".jpg")
```
