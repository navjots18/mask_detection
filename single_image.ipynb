```python
import uuid
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
import cv2
import requests
from io import BytesIO
```

```python
url='https://encrypted-tbn0.gstatic.com/images?q=tbn%3AANd9GcTq_P5U_rudEUUk2UfJIeYh1lugZ9-oIhyXGg&usqp=CAU'
rel = requests.get(url)
rel.status_code
```

```python
f = open("test.jpg","wb")
f.write(rel.content)
f.close()
```

```python
frame = cv2.imread("test.jpg")
```

```python
plt.imshow(frame)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
faces= classifier.detectMultiScale(frame)
pil = Image.open("test.jpg") 
arr= np.array(pil)
face=faces[0]
```

```python
len(faces)
```

```python
x,y,h,w = face
img = Image.fromarray(arr[y:y+h,x:x+w,], 'RGB')
img = img.resize((40,40))
plt.figure()
plt.imshow(arr[y:y+h,x:x+w,])
img = img.resize((40,40))
```

```python
img = img.save("test.jpg")
```
